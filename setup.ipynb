```python
import cv2
import mss
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

sct = mss.mss()



SCREEN_WIDTH = 636
SCREEN_HEIGHT = 424

#! Adjust me to crop the screen, adjust the numbers for the proper crop to show up
SCREEN_X = 962
SCREEN_Y = 536

#! Adjust me to change which monitor is captured
monitor_number = 1

cur_monitor = sct.monitors[monitor_number]

monitor = {"top": cur_monitor["top"] + SCREEN_Y, 
           "left": cur_monitor["left"] + SCREEN_X, 
           "width": SCREEN_WIDTH, 
           "height": SCREEN_HEIGHT,
           "mon": monitor_number}

def grab_screen_shot():
        sct_img = sct.grab(monitor)
        frame = cv2.cvtColor(np.asarray(sct_img, dtype=np.uint8), cv2.COLOR_RGB2BGR)
        return frame

screen = grab_screen_shot()

plt.imshow(screen)
```
